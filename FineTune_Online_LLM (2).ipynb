```python
config = {
    "hugging_face_username": "Shekswess",
    "model_config": {
        "base_model": "unsloth/llama-3-8b-Instruct-bnb-4bit",  # Base model
        "finetuned_model": "llama-3-8b-Instruct-bnb-4bit-medical",  # Fine-tuned model name
        "max_seq_length": 2048,  # Maximum sequence length
        "dtype": torch.float16,  # Data type
        "load_in_4bit": True,  # Load model in 4-bit
    },
    "lora_config": {
        "r": 16,  # LoRA layers
        "target_modules": ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],  # Target modules
        "lora_alpha": 16,  # LoRA alpha
        "lora_dropout": 0,  # LoRA dropout
        "bias": "none",  # Bias
        "use_gradient_checkpointing": True,  # Gradient checkpointing
        "use_rslora": False,  # Use RSLora
        "use_dora": False,  # Use DoRa
        "loftq_config": None,  # LoFTQ config
    },
    "training_dataset": {
        "name": "Shekswess/medical_llama3_instruct_dataset_short",  # Dataset name
        "split": "train",  # Dataset split
        "input_field": "prompt",  # Input field
    },
    "training_config": {
        "per_device_train_batch_size": 2,  # Batch size
        "gradient_accumulation_steps": 4,  # Gradient accumulation steps
        "warmup_steps": 5,  # Warmup steps
        "max_steps": 0,  # Max steps (0 if using epochs)
        "num_train_epochs": 1,  # Number of training epochs
        "learning_rate": 2e-4,  # Learning rate
        "fp16": not torch.cuda.is_bf16_supported(),  # Use fp16 if bf16 not supported
        "bf16": torch.cuda.is_bf16_supported(),  # Use bf16 if supported
        "logging_steps": 1,  # Logging steps
        "optim": "sgd",  # Use SGD optimizer
        "weight_decay": 0.01,  # Weight decay
        "lr_scheduler_type": "linear",  # Learning rate scheduler
        "seed": 42,  # Seed
        "output_dir": "outputs",  # Output directory
    }
}
```

```python
base_model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=config.get("model_config").get("base_model"),
    max_seq_length=config.get("model_config").get("max_seq_length"),
    dtype=config.get("model_config").get("dtype"),
    load_in_4bit=config.get("model_config").get("load_in_4bit"),
)
```

```python
model = FastLanguageModel.get_peft_model(
    base_model,
    r=config.get("lora_config").get("r"),
    target_modules=config.get("lora_config").get("target_modules"),
    lora_alpha=config.get("lora_config").get("lora_alpha"),
    lora_dropout=config.get("lora_config").get("lora_dropout"),
    bias=config.get("lora_config").get("bias"),
    use_gradient_checkpointing=config.get("lora_config").get("use_gradient_checkpointing"),
    random_state=42,
    use_rslora=config.get("lora_config").get("use_rslora"),
    use_dora=config.get("lora_config").get("use_dora"),
    loftq_config=config.get("lora_config").get("loftq_config"),
)
```

```python
dataset_train = load_dataset(config.get("training_dataset").get("name"), split=config.get("training_dataset").get("split"))
```

```python
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset_train,
    dataset_text_field=config.get("training_dataset").get("input_field"),
    max_seq_length=config.get("model_config").get("max_seq_length"),
    dataset_num_proc=2,
    packing=False,
    args=TrainingArguments(
        per_device_train_batch_size=config.get("training_config").get("per_device_train_batch_size"),
        gradient_accumulation_steps=config.get("training_config").get("gradient_accumulation_steps"),
        warmup_steps=config.get("training_config").get("warmup_steps"),
        max_steps=config.get("training_config").get("max_steps"),
        num_train_epochs=config.get("training_config").get("num_train_epochs"),
        learning_rate=config.get("training_config").get("learning_rate"),
        fp16=config.get("training_config").get("fp16"),
        bf16=config.get("training_config").get("bf16"),
        logging_steps=config.get("training_config").get("logging_steps"),
        optim=config.get("training_config").get("optim"),  # SGD optimizer
        weight_decay=config.get("training_config").get("weight_decay"),
        lr_scheduler_type=config.get("training_config").get("lr_scheduler_type"),
        seed=42,
        output_dir=config.get("training_config").get("output_dir"),
    ),
)
```

```python
gpu_statistics = torch.cuda.get_device_properties(0)
reserved_memory = round(torch.cuda.max_memory_reserved() / 1024**3, 2)
max_memory = round(gpu_statistics.total_memory / 1024**3, 2)
print(f"Reserved Memory: {reserved_memory}GB")
print(f"Max Memory: {max_memory}GB")
```

```python
trainer_stats = trainer.train()
```

```python
used_memory = round(torch.cuda.max_memory_allocated() / 1024**3, 2)
used_memory_lora = round(used_memory - reserved_memory, 2)
used_memory_percentage = round((used_memory / max_memory) * 100, 2)
used_memory_lora_percentage = round((used_memory_lora / max_memory) * 100, 2)

print(f"Used Memory: {used_memory}GB ({used_memory_percentage}%)")
print(f"Used Memory for training (fine-tuning) LoRA: {used_memory_lora}GB ({used_memory_lora_percentage}%)")
```

```python
# List of questions without repeating "system" or "user"
questions = [
    "Can you provide an overview of lung squamous cell carcinoma?",
    "What are the symptoms of asthma?",
    "How is tuberculosis diagnosed?"
]

# Properly structured prompts
prompts = [f"system: You are a medical professional. Answer the following question truthfully.\nuser: {q}" for q in questions]

# Tokenize multiple inputs with correct structure
inputs = tokenizer(
    prompts,
    return_tensors="pt",
    padding=True,  # Ensures all inputs are padded to the same length
    truncation=True  # Truncates if input exceeds max length
).to("cuda")

# Generate responses for each input prompt
outputs = model.generate(**inputs, max_new_tokens=256, use_cache=True)

# Decode the generated responses
decoded_outputs = tokenizer.batch_decode(outputs, skip_special_tokens=True)

# Print each question with its corresponding answer
for question, response in zip(questions, decoded_outputs):
    print(f"Question: {question}")
    print(f"Answer: {response.strip()}\n")
```

```python
@anvil.server.callable
def answer(question):
  questions = []
  questions.append(question)
  responses = []
  # Properly structured prompts
  prompts = [f"system: You are a medical professional. Answer the following question truthfully.\nuser: {q}" for q in questions]

  # Tokenize multiple inputs with correct structure
  inputs = tokenizer(
      prompts,
      return_tensors="pt",
      padding=True,  # Ensures all inputs are padded to the same length
      truncation=True  # Truncates if input exceeds max length
  ).to("cuda")

  # Generate responses for each input prompt
  outputs = model.generate(**inputs, max_new_tokens=256, use_cache=True)

  # Decode the generated responses
  decoded_outputs = tokenizer.batch_decode(outputs, skip_special_tokens=True)

  for question,response in zip(questions, decoded_outputs):
    responses.append(f"Answer: {response.strip()}\n")

  return responses
```
